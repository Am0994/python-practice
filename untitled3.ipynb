#Global variables

```python
x = "awesome"

def myfunc():
  print("Python is " + x)

myfunc()
```

```python
x = "awesome"

def myfunc():
  x = "fantastic"
  print("Python is " + x)

myfunc()

print("Python is " + x)
```

```python
def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```
